```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets
```

```python
X, y, coef = datasets.make_regression(
    n_samples=500,
    n_features=1,
    n_informative=1,
    noise=20,
    coef=True,
    random_state=0,
)
X+=abs(round(np.min(X)))
y+=abs(round(np.min(y)))
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
x_min, x_max = round(np.min(X)), round(np.max(X))
x_line = np.arange(x_min, x_max, 0.5)
y_line = x_line * coef
```

```python
plt.figure(figsize=(8,5))
plt.scatter(X, y)
plt.plot(x_line, y_line, "-r", lw=3, label=f"y = {int(coef)}x + 0")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc="upper left")
axe = plt.gca()
axe.set_xlim([0,6])
axe.set_ylim([0, 300])
plt.show()
```

```python
ms = [10, 30, 50]
bs = [0, 0, 0]
styles = ["-r", "-b", "-g"]
y_lines = []
for m in ms:
    line = x_line * m
    y_lines.append(line)
```

```python
plt.figure(figsize=(8,5))
for ys, coef, st in zip(y_lines, ms, styles):
    plt.plot(x_line, ys, st, lw=3, label=f"y = {int(coef)}x + 0")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc="upper left")
axe = plt.gca()
axe.set_xlim([0,6])
axe.set_ylim([0, 300])
plt.show()
```

```python
ms = [45, 45, 45]
bs = [-20, 40, 120]
styles = ["-r", "-b", "-g"]
y_lines = []
for b in bs:
    line = x_line * 45 + b
    y_lines.append(line)
```

```python
plt.figure(figsize=(8,5))
for ys, b, st in zip(y_lines, bs, styles):
    plt.plot(x_line, ys, st, lw=3, label=f"y = 45x {'+' if b >= 0 else '-'} {abs(b)}")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc="upper left")
axe = plt.gca()
axe.set_xlim([0,6])
axe.set_ylim([0, 300])
plt.show()
```

```python
xs = np.linspace(0, 6, 100)
ys_quad = (xs - 3)**2
ys_deriv = 2 * (xs - 3)
plt.figure(figsize=(8,5))
plt.plot(xs, ys_quad, "-r", label="Quadratic function")
plt.plot(xs, ys_deriv, "-b", label="Derivative")
plt.axhline(0, color="black")
plt.axvline(x = 3, color = 'g', label="Optimum marker")
axe = plt.gca()
axe.set_xlim([0,6])
axe.set_ylim([-6, 9])
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc="upper left")
plt.show()
```

```python
import matplotlib.patches as p

def quadratic(x):
    return (x - 3)**2

def derivative(x):
    return 2*(x - 3)

xs = np.linspace(0, 6, 100)
ys_quad = quadratic(xs)
point_1 = (1, 4)
target = (3, 0)

slope = derivative(point_1[0])
y = quadratic(point_1[0])
deriv_xs = (point_1[0]- 0.5, point_1[0] + 0.5)
deriv_ys = (y-(slope*0.5), y+(slope*0.5))


plt.figure(figsize=(8,5))
plt.plot(xs, ys_quad, "-r", label="Quadratic function")

plt.plot(deriv_xs, deriv_ys, "-g", linewidth=3, label="Slope")

arrow = p.FancyArrowPatch(point_1, target, mutation_scale=20, arrowstyle="->")
plt.scatter([point_1[0]], [point_1[1]], color="blue", zorder=5)
plt.scatter([3], [0], color="blue", zorder=5)
axe = plt.gca()
axe.add_patch(arrow)
axe.set_xlim([0,6])
axe.set_ylim([0, 9])
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(loc="upper left")
plt.show()
```

```python
# Difference vis
xs = [10, 11, 20, 30]
ys = [28, 27, 46, 73]
y_hats = [30, 32, 50, 70]

def fun(x):
    return x * 2 + 10

x_line = np.linspace(1, 100, 1000)
y_line = fun(x_line)

# errors
errors = []
for x, y, y_hat in zip(xs, ys, y_hats):
    line = ([x, x], [y, y_hat])
    errors.append(line)

plt.figure(figsize=(8,5))
plt.xlabel("X")
plt.ylabel("Y")
plt.scatter(xs, ys)
plt.plot(x_line, y_line, "-g", label="y = 2x + 10", linewidth=3)
for error in errors:
    if error == errors[0]:
        plt.plot(error[0], error[1], "-r", linewidth=3, label="Error")
    else:
        plt.plot(error[0], error[1], "-r", linewidth=3)
axe = plt.gca()
axe.set_xlim([8, 35])
axe.set_ylim([20, 80])
plt.legend(loc='upper left')
plt.show()
```
