### What is Linear Regression?

```python
#@title
import numpy as np
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt

bias = 100
x, y, coef = make_regression(n_samples=500, n_features=1, n_informative=1, noise=15, bias=bias, random_state=1, coef=True)
min = round(np.min(x))
x -= min 

plt.figure(figsize=(16, 9))
plt.scatter(x, y, label='Data points')
plt.legend(loc='lower right')
plt.xlabel('X')
plt.ylabel('Y')
axis = plt.gca()
axis.set_ylim([0, 200])
plt.show()
```

```python
#@title
bias = 100
x, y, coef = make_regression(n_samples=500, n_features=1, n_informative=1, noise=15, bias=bias, random_state=1, coef=True)
min = round(np.min(x))
x -= min 
x_line = np.arange(0, 7)
y_line = bias + (x_line + min) * coef

plt.figure(figsize=(16, 9))
plt.scatter(x, y, label='Data points')
plt.plot(x_line, y_line, '-r', label="Regression line", lw=4)
plt.legend(loc='lower right')
plt.xlabel('X')
plt.ylabel('Y')
axis = plt.gca()
axis.set_ylim([0, 200])
plt.show()
```

### **Line Equation:**


\begin{align}
        \mathbf{y = m \cdot x + b} \\
        m = slope \\
        b = bias
\end{align}

```python
#@title
bias = 100
x, y, coef = make_regression(n_samples=500, n_features=1, n_informative=1, noise=15, bias=bias, random_state=1, coef=True)
min = round(np.min(x))
x -= min 
x_line = np.arange(0, 7)
biasies = [40,70, 100, 20]
coefs = [25, 15, 0, 15]
colors = ['-r', '-y', '-g', '-m']
labels = []
y_lines = []
for bias, coef in zip(biasies, coefs):
    y_line = bias + (x_line ) * coef
    y_lines.append(y_line)
    labels.append(f"y = {coef}x + {bias}")

plt.figure(figsize=(16, 9))
plt.scatter(x, y, label='Data points')
for idx in range(len(coefs)):
    plt.plot(x_line, y_lines[idx], colors[idx], label=labels[idx], lw=4)
plt.legend(loc='upper left')
plt.xlabel('X')
plt.ylabel('Y')
axis = plt.gca()
axis.set_ylim([0, 200])
plt.show()
```

### **How god/bad is the line?**

```python
#@title
bias = 100
x, y, coef = make_regression(n_samples=500, n_features=1, n_informative=1, noise=15, bias=bias, random_state=1, coef=True)
min = round(np.min(x))
x -= min 
x_line = np.arange(0, 7)
bias = 50
coef = 20
def y_line_func(xs, coef, bias):
    y_line = bias + (xs) * coef
    return y_line
y_line = y_line_func(x_line, coef, bias)

plt.figure(figsize=(16, 9))
plt.scatter(x, y, label='Data points')
plt.plot(x_line, y_line, "-g", label=f"y = {coef}x + {bias}", lw=4)

# Error lines
np.random.seed(50)
gets = np.random.randint(low=0, high=500, size=20)
xs = x[gets]
ys = y[gets]
ys_line = y_line_func(xs, coef, bias)

for ex_x, ex_y, pred_y in zip(xs, ys, ys_line):
    x = np.array([ex_x, ex_x])
    y = np.array([ex_y, pred_y[0]])
    plt.plot(x, y, "-r", lw=4)
plt.legend(loc='upper left')
plt.xlabel('X')
plt.ylabel('Y')
axis = plt.gca()
axis.set_ylim([0, 200])
plt.show()
```

### **Error function(Mean Squared Error):**


\begin{align}
    &\mathrm{\theta = \frac{1}{n} \sum_{i=0}^{n} (\hat{y_i} - y_i)^2} \\
    &\mathrm{\hat{y} = 20x + 50}
\end{align}

```python
xs = np.linspace(-2, 2, 1000)
ys = xs**2
plt.figure(figsize=(16,9))
plt.plot(xs, ys, label="y = x^2")
plt.legend(loc="upper left")
plt.show()
```
